## Project Overview

### Project Title: Global Cybersecurity Threat Analysis Using Python

This project performs an exploratory data analysis (EDA) on a real-world cybersecurity dataset. The primary objective is to understand the trends and patterns of cyber threats across different countries, industries, and attack types.

After preprocessing, several visualizations are created using Plotly and Seaborn to gain meaningful insights. These include histograms, box plots, pie charts, area charts, and line graphs that illustrate relationships between countries, attack types, target industries, financial losses, affected users, security vulnerabilities, defense mechanisms, and incident resolution time.

The project helps identify which industries are most frequently targeted, the financial impact of cyberattacks, the effectiveness of defense mechanisms, and trends in cybersecurity incidents. These insights can assist organizations in improving cybersecurity strategies and risk management.

### Project Summary:

This project analyzes global cybersecurity threats using Python-based data analysis libraries. It explores different aspects of cyberattacks, including attack types, target industries, countries, financial losses, attack sources, and security vulnerabilities. Through data cleaning, statistical analysis, and interactive visualizations, the project uncovers important trends and patterns in cybersecurity incidents. The analysis demonstrates how data science techniques can be applied to cybersecurity to support better decision-making, enhance threat awareness, and improve organizational security planning.

```python
df=pd.read_csv("Global_Cybersecurity_Threats.csv")
```

```python
fix=px.histogram(df,x="Country",y="Year",color="Country",title="Histogram chart of Country and Year ")
fix.show()
```

### Purpose of Histogram

Shows cyber attack records for different countries.

#### Interpretation
1. Each colored bar represents a country.
2. Helps compare the distribution of attack records among countries.

### Box chart 

```python
fix=px.box(df,x="Attack Type",y="Country",color="Attack Type",title="Box Chart for Cyber Attack according to Country ")
fix.show()
```

### Purpose of Box chart

Compares attack types across different countries.

#### Interpretation
1. Shows how attack types vary between countries.
2. Helps identify variations and unusual observations.

### Pie Chart for checking the Target Industries

```python
count=df["Target Industry"].value_counts().reset_index()
count


```

```python
fix=px.pie(count,names="Target Industry",values="count",title="Pie Chart for Industries Targets by Hackers")
fix
```

### Purpose

Shows the percentage of attacks on each industry.

#### Interpretation
Each slice represents one industry.
Larger slices indicate industries targeted more frequently.

#### Example:

1. Healthcare – 22%
2. Finance – 18%
3. IT – 16%

### After that compare Target industry with Financial Loss (in Million $)

```python
fix=px.histogram(df,x="Target Industry",y="Financial Loss (in Million $)",color="Target Industry",title="Histogram Chart of Financial loss Over Countries")
fix.show()
```

### Interpretation

### Chart for Attack Source and peoples affected by that

```python
fix=px.area(df.head(1000),y="Attack Source",x="Number of Affected Users",color="Attack Source",title="Area chart for Attack Source over No. of Affected Users")
fix.show()
```

### Purpose

Shows the relationship between attack source and affected users.

#### Interpretation
1. Larger shaded areas represent attacks affecting more users.
2. Useful for understanding how attack sources impact users.
